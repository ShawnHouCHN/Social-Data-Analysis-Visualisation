**How does K-nearest-neighbors (KNN) work? Explain in your own words.**

KNN works by the assumption that data points that lie near to each other have a high propability to be similar. Requirements for doing KNN you need a way to calculate the distance between data points (so you are able to determine the approximity), some supervised examples (so you have an idea of what you are classifing for) and you need a value for K (k needs to be a possitive integer). A low k value leads to complex classes (many small classes). A large k on the other hand leads to smoother and bigger classes

KNN wont work for datasets of many dimensions, since the more dimensions the easier it is for two datapoints to be distant. This is shown in figure 12-6 where the smallest distance divided by the average distance approaches 1, when the number of dimensions approaches infinity

```python
import geoplotlib
from geoplotlib.utils import read_csv, BoundingBox, DataAccessObject
import pandas as pd

with open("C:\Users\Simon\Documents\GitHub\Social-Data-Analysis-Visualisation\Simon\Week03\SFPD_Incidents_-_from_1_January_2003.csv", 'rb') as f:
    df = pd.read_csv(f)
    
focuscrimes = ['PROSTITUTION', 'DRUG/NARCOTIC', 'DRIVING UNDER THE INFLUENCE']
focus_df = df[df.Category.isin(focuscrimes)]
focus_df = focus_df[focus_df.Y != 90] #there is one freaking outlier fucking it up for everyone else..
```

```python
lon = focus_df['X'].tolist()
lat = focus_df['Y'].tolist()
lon_max = max(lon)
lon_min = min(lon)
lat_max = max(lat)
lat_min = min(lat)

geo_data_for_plotting = {"lat": lat, "lon": lon}
bbox = BoundingBox(north=lat_max, west=lon_min, south=lat_min, east=lon_max)
geoplotlib.set_bbox(bbox)
geoplotlib.kde(geo_data_for_plotting, bw=5, cut_below=1e-4)
geoplotlib.inline()
```

```python
import numpy as np
import pylab as pl
from sklearn import neighbors, datasets

#renaming data
X = lon
Y = lat

h = .02 # step size in the mesh
knn=neighbors.KNeighborsClassifier()
knn.fit(X, Y)

# Plot the decision boundary. For that, we will asign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])


# Put the result into a color plot
Z = Z.reshape(xx.shape)
pl.figure(1, figsize=(4, 3))
pl.set_cmap(pl.cm.Paired)
pl.pcolormesh(xx, yy, Z)

# Plot also the training points
pl.scatter(X[:,0], X[:,1],c=Y )
pl.xlabel('Sepal length')
pl.ylabel('Sepal width')

pl.xlim(xx.min(), xx.max())
pl.ylim(yy.min(), yy.max())
pl.xticks(())
pl.yticks(())

pl.show()
```
