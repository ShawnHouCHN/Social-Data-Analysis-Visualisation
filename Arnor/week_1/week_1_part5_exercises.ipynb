- Q4. Create your own version of Figure 3-9 and add to the notebook. Explain in your own words why 3-9 is a better representation of the data than 3-8.

```python
no_cars_made_q1 = [96, 65, 93, 74, 82]
no_cars_made_q2 = [93, 60, 83, 73, 85]

plt.scatter(no_cars_made_q1, no_cars_made_q2)

plt.title("Car sales in Q1 and Q2")
plt.xlabel("Number of cars made in Q1")
plt.ylabel("Number of cars made in Q2")
plt.axis("equal")

plt.show()
```

Having the axes scaled differently means that the visually apparent variance is misleading as the distance between data points is not on the same scale. That is, both axes might seem to exhibit the same amount of variance but the reason for that is that one of them has been "stretched out" by having a lower scale. Having the axes on the same scale gives a more realistic comparison of the variance.
