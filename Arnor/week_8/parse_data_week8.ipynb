```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("/Users/Work/programming/repositories/github_data/SAL_data/SFPD_Incidents_-_from_1_January_2003.csv")
df = df.dropna()
df = df[(np.abs(stats.zscore(df[['X','Y']])) < 5).all(axis=1)]
```

```python
df.head()
```

```python
PR_locs = df[df['Category'] == 'PROSTITUTION'][['X', 'Y']].copy()
PR_locs.head()
```

```python
from sklearn import cluster

kmeans_data = np.array([PR_locs['X'], PR_locs['Y']]).T
kmeans_cluster_n = range(2,7)

cluster_centers = []

for n in kmeans_cluster_n:
    
    kmeans = cluster.KMeans(n_clusters=n)
    kmeans.fit(kmeans_data)
    cluster_centers.append(kmeans.cluster_centers_)
    
    PR_locs['k' + str(n)] = kmeans.labels_
```

```python
PR_locs.head()
```

```python
PR_locs.memory_usage()
```

```python
PR_locs = PR_locs.sort_values(['k2', 'k3', 'k4', 'k5', 'k6'], ascending=True)
PR_locs
```

```python
PR_locs.to_json('kmeansPR-values.json', orient='values')
```
